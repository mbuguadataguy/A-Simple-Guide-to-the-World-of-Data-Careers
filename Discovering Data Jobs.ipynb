```python
#Find a dataset with the 2 letter country abbreviation
# I found one from this source(https://www.allbusinesstemplates.com/template/U2FC8/2-letter-codes-country-abbreviationsxlsx/)
#import it
df2 = pd.read_excel(r"D:\JUDE PROGRAMMING\PORTFOLIO PROJECTS\DATA SCIENCE SALARIES EDA\allcountryabbreviations.xlsx")
```

```python
df2
```

```python
#Drop some columns
df2 = df2.drop(columns = 'Unnamed: 0',)
df2 
```

```python
#Drop first row
df2 = df2.drop(index = 0,axis  =1)
df2 
```

```python
#Rename columns
df2 = df2.rename(columns = {'Alphabetically by abbreviation':'employee_residence','Unnamed: 2':'country'})
```

```python
#Capitalize the employee_residence column
df2['employee_residence'] = df2['employee_residence'].str.upper()
```

```python
df2
```

```python
#Merge df2 and df
pd.merge(df,df2,how="left", on=["employee_residence"])
```

```python
#Create new dataset

df3 = pd.merge(df,df2,how="left", on=["employee_residence"])
```

```python
df3.head(50)
```

```python
df3['country'].value_counts().head(5).plot(kind ='barh',title = 'Location Of Data Related Jobs'.upper(),xlabel ='Number of Jobs',ylabel = 'Countries',color ='teal')
```

```python
#The United Kingdom column is too messy. I did not spot it when cleaning the dataset.Fix it

df3['country'] = df3['country'].str.replace('United Kingdom (no new registrations, see also UK)', 'United Kingdom')
```

```python
df3['country'].value_counts().head(5).plot(kind ='barh',title = 'Location Of Unique Data Science Jobs'.upper(),xlabel ='Number of Jobs',ylabel = 'Countries',color ='teal')
```
